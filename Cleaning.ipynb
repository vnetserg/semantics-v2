```python
all_clusters = set()
data_cl = data.copy()
for group in CLUSTER_GROUPINGS:
    for cluster in group:
        if cluster in all_clusters:
            raise ValueError(cluster)
            all_clusters.add(cluster)
        data_cl.replace(cluster, group[0], inplace=True)
    #data_cl.replace(group[:1], group[0], inplace=True)
```
